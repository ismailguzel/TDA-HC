```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
iris = sns.load_dataset("iris")
X = iris.drop('species', axis=1)
y = iris.species
```

```python
model = KMeans(n_clusters=4, random_state=42)
y_pred = model.fit_predict(X)
```

```python
X= X.to_numpy()
```

```python
fig, ax = plt.subplots(1,2)
fig.set_size_inches(12,4)
sns.scatterplot(x=X[:,0], y=X[:,1],hue = iris.species, legend=True, ax=ax[0]);
sns.scatterplot(x=X[:,0], y=X[:,1],hue = y_pred, legend=True, ax=ax[1]);
```

```python
fig, ax = plt.subplots(1,2)
fig.set_size_inches(12,4)
sns.scatterplot(x=X[:,0], y=X[:,2],hue = iris.species, legend=True, ax=ax[0]);
sns.scatterplot(x=X[:,0], y=X[:,2],hue = y_pred, legend=True, ax=ax[1]);
```

```python
from sklearn.metrics import silhouette_score, rand_score, v_measure_score
```

```python
model1 = KMeans(n_clusters=4, random_state=0)
model1.fit_predict(X)
model2 = KMeans(n_clusters=3, random_state=0)
model2.fit_predict(X)
```

```python
silhouette_score(X,model1.labels_,)
```

```python
silhouette_score(X,model2.labels_)
```

```python
rand_score(y, model1.labels_)
```

```python
rand_score(y, model2.labels_)
```

```python
v_measure_score(y, model1.labels_)
```

```python
v_measure_score(y, model2.labels_)
```
